# Theory


1. What is the difference between a function and a method in Python?
* A function is an independent block of code that can be called from anywhere, while a method is a function tied to a class or object and must be called on that specific object or within that class.
* Example: A function can be defined as: def greet(name): return 'Hello, ' + name, while a method inside a class might look like: class Person: def greet(self): return 'Hello!'. Understanding this distinction is crucial for leveraging object-oriented programming effectively.
2. Explain the concept of function arguments and parameters in Python.
* Parameters are the named variables in a function’s definition, whereas arguments are the actual values passed to the function when it is called.      
* Example: In the function definition def add(x, y): return x + y, x and y are parameters, and calling add(3, 4) passes 3 and 4 as arguments. Recognizing the difference aids in writing flexible and reusable functions.
3. What are the different ways to define and call a function in Python?
* You can define functions using the standard def keyword or by using a lambda expression for simple, one-line functions.
* Example: A standard function: def square(x): return x * x. Calling square(5) returns 25. A lambda function example is square_lambda = lambda x: x * x, and calling square_lambda(5) also returns 25. This variety offers flexibility in function creation, suiting different coding scenarios.
4. What is the purpose of the `return` statement in a Python function?
* The return statement terminates function execution and sends a result back to the caller.
* Example: In the function def multiply(a, b): return a * b, calling multiply(4, 5) will return 20. Utilizing return appropriately ensures that functions can communicate outcomes effectively.
5. What are iterators in Python and how do they differ from iterables?
* An iterable is any Python object capable of returning its elements one at a time (like a list) by implementing the __iter__() method, while an iterator is an object that represents a stream of data and implements both __iter__() and __next__() methods.
* Example: Given a list such as numbers = [1, 2, 3], obtaining an iterator by calling iter(numbers) allows you to retrieve elements one at a time with next(iterator), which will return 1 first. Differentiating these helps in writing more efficient loops and data processing routines.
6. Explain the concept of generators in Python and how they are defined.
* Generators are a special kind of iterator that yield items one at a time using the yield keyword instead of returning all values at once. They are defined like normal functions but use yield to produce a sequence of values.
* Example: A generator might be defined as: def count_up(n): i = 0; while i < n: yield i; i += 1. Iterating through count_up(3) would yield 0, then 1, then 2. Generators provide a powerful way to work with large or infinite sequences efficiently.
7. What are the advantages of using generators over regular functions?
* Generators generate items on the fly, making them memory efficient, and use lazy evaluation, computing values as needed which can save time and resources.
* Example: A generator to read a huge file line by line can be defined as: def read_large_file(file_path): with open(file_path, 'r') as f: for line in f: yield line. These benefits make generators a go-to solution when handling data-intensive tasks.
8. What is a lambda function in Python and when is it typically used?
* A lambda function is an anonymous, small function defined using the lambda keyword. It is typically used for short, throwaway functions, especially as arguments to higher-order functions.
* Example: A lambda function to add two numbers can be defined as add = lambda a, b: a + b; calling add(2, 3) returns 5. Lambda functions help keep the code concise and are ideal for quick, on-the-fly computations.
9. Explain the purpose and usage of the `map()` function in Python.
* The map() function applies a given function to every item of an iterable and returns an iterator with the results.
* Example: To double each number in a list, you might write numbers = [1, 2, 3, 4] and then use map with a lambda function as doubled = list(map(lambda x: x * 2, numbers)), which results in [2, 4, 6, 8]. The map() function streamlines the process of applying transformations across data sets.
10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
* The map() function transforms each element by applying a function, the filter() function selects elements that satisfy a given condition, and the reduce() function applies a function cumulatively to the items
 of an iterable to reduce it to a single value (available from the functools module).
* Example: For a list numbers = [1, 2, 3, 4]: Using map to multiply each element by 2 gives [2, 4, 6, 8]. Using filter with a condition for even numbers gives [2, 4]. Using reduce (from functools import reduce) to sum numbers gives 10. These functions offer different paradigms for data transformation and aggregation in Python.

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list
def sum_of_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

nums = [1, 2, 3, 4, 5, 6]
print(sum_of_even_numbers(nums))
```

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def square_list(nums):
    return [num ** 2 for num in nums]

numbers = [1, 2, 3, 4, 5]
print(square_list(numbers))
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

prime_numbers = [num for num in range(1, 201) if is_prime(num)]
print(prime_numbers)
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration

        if self.count == 0:
            self.count += 1
            return self.a

        if self.count == 1:
            self.count += 1
            return self.b

        next_fib = self.a + self.b
        self.a, self.b = self.b, next_fib
        self.count += 1
        return next_fib

# Example usage
fib_iterator = FibonacciIterator(10)
print(list(fib_iterator))

```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage
for power in powers_of_2(5):
    print(power)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        for line in file:
            yield line.strip()  # Remove trailing newline characters

for line in read_file_line_by_line("/content/test.txt"):
    print(line)
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
data = [(1, 5), (3, 2), (4, 8), (2, 1)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

celsius_list = [0, 10, 25, 40]
fahrenheit_list = list(map(celsius_to_fahrenheit, celsius_list))
print(fahrenheit_list)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(input_string):
    vowels = set("aeiouAEIOU")
    result = map(lambda ch: "" if ch in vowels else ch, input_string)
    return "".join(result)
test_string = "Hello World!"
print(remove_vowels(test_string))
```
